Based on the screenshot, the error you are seeing (TypeError: ... takes 1 positional argument but 2 were given) is not actually because the number is negative. It is happening because of how you are passing the number.

In Python classes, self counts as the first argument.

Your definition: def printNumber(self): expects only 1 argument (self).

Your call: obj.printNumber(-5) passes 2 arguments (self and -5).

Here are the two ways to fix this, depending on what the specific coding task requires:

Option 1: Pass the number as an Argument (Matches your screenshot)
If you want to run the code exactly like obj.printNumber(-5), you must update your function definition to accept a variable n.

```python
class Solution:
    # Update 'self' to 'self, n' to accept the number you pass in
    def printNumber(self, n):
        print(n)

# Now this will work:
obj = Solution()
obj.printNumber(-5)
```

```python
obj.printNumber(7)
```
