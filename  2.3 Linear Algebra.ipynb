### (column vector) × (row vector) → matrix (Outer product )


```python
col = torch.tensor([[1],
                    [2],
                    [3]])     # shape (3, 1)

row = torch.tensor([[4, 5, 6]])  # shape (1, 3)
```

```python
M = col @ row   # Outer product
print(M)
```

✔️ Key Points

A row vector in PyTorch: shape (1, n)

A column vector: shape (m, 1)

Their matrix product (m,1) @ (1,n) gives a matrix of shape (m,n)

```python
A = torch.tensor([[1., 2., 3.],
                  [4., 5., 6.]])  # (2, 3)

x = torch.tensor([[1.],
                  [1.],
                  [1.]])          # (3, 1)

y = A @ x
print(y)
```

⚙️ 2. Matrix × Row Vector (NOT standard)

This is not a standard linear algebra operation unless you transpose the vector.

If you have a row vector:

```python
v = torch.tensor([1., 1., 1.])   # shape (3,)
v.shape
```

```python
A @ v
```

```python
v_col = v.reshape(3, 1)
A @ v_col
```

| Operation     | Shapes        | Result |
| ------------- | ------------- | ------ |
| A @ x         | (m,n) × (n,1) | (m,1)  |
| xᵀ @ A        | (1,n) × (n,m) | (1,m)  |
| Outer product | (m,1) × (1,n) | (m,n)  |
