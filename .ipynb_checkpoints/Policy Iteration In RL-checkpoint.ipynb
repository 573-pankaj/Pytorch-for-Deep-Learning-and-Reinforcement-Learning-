```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
# ==========================================
# 1. THE ENVIRONMENT (Grid World)
# ==========================================
class GridWorld:
    def __init__(self, size=4):
        self.size = size
        self.n_states = size * size
        self.n_actions = 4 # 0: Up, 1: Right, 2: Down, 3: Left
        self.goal_state = self.n_states - 1 # Bottom-right corner
        
        # Define obstacles (optional, for flavor)
        # For a 4x4 grid, let's put a "hole" at index 5 and 12
        self.holes = [5, 12] 
        
        # Transition Probabilities: P(s' | s, a)
        # We model a "slippery" world: 
        # 80% chance to move intended direction, 
        # 10% chance to slip left, 10% chance to slip right relative to action
        self.transitions = {} 
        self._build_transitions()

    def _get_coords(self, s):
        return s // self.size, s % self.size

    def _get_state(self, r, c):
        # Clip coordinates to grid boundaries
        r = max(0, min(r, self.size - 1))
        c = max(0, min(c, self.size - 1))
        return r * self.size + c

    def _build_transitions(self):
        # Action effects: (d_row, d_col)
        # 0: Up, 1: Right, 2: Down, 3: Left
        moves = [(-1, 0), (0, 1), (1, 0), (0, -1)]
        
        for s in range(self.n_states):
            self.transitions[s] = {}
            for a in range(self.n_actions):
                self.transitions[s][a] = []
                
                # If terminal state (Goal or Hole), interaction ends
                if s == self.goal_state or s in self.holes:
                    self.transitions[s][a].append((1.0, s, 0, True))
                    continue

                # Possible dynamics (Intended, Slip 1, Slip 2)
                # Intended action
                next_r, next_c = np.array(self._get_coords(s)) + np.array(moves[a])
                s_prime = self._get_state(next_r, next_c)
                
                # Reward: +1 for Goal, -1 for Hole, -0.01 for step (living cost)
                if s_prime == self.goal_state: r = 1.0
                elif s_prime in self.holes: r = -1.0
                else: r = 0.0
                
                # We simply simulate a deterministic move for clarity in this basic version,
                # but Policy Iteration works perfectly for stochastic too.
                # Format: (probability, next_state, reward, is_terminal)
                self.transitions[s][a].append((1.0, s_prime, r, False))


```

```python
# ==========================================
# 2. POLICY ITERATION ALGORITHM
# ==========================================

def policy_evaluation(env, policy, V, gamma=0.9, theta=1e-6):
    """
    Step 1: Calculate the Value Function V(s) for a specific policy.
    We iterate until the values stop changing (convergence).
    """
    while True:
        delta = 0
        for s in range(env.n_states):
            v_old = V[s]
            
            # Bellman Expectation Equation:
            # V(s) = sum( P(s'|s,a) * [r + gamma * V(s')] )
            a = policy[s] # The action our current policy tells us to take
            
            v_new = 0
            for prob, s_prime, reward, done in env.transitions[s][a]:
                # If done, there is no future value (V(s') = 0)
                future_val = 0 if done else V[s_prime]
                v_new += prob * (reward + gamma * future_val)
            
            V[s] = v_new
            delta = max(delta, abs(v_old - v_new))
        
        if delta < theta:
            break
    return V

def policy_improvement(env, V, policy, gamma=0.9):  # <--- Added 'policy' argument here
    """
    Step 2: Make the policy 'greedy'.
    """
    policy_stable = True
    new_policy = np.zeros(env.n_states, dtype=int)
    
    for s in range(env.n_states):
        old_action = policy[s] # Now 'policy' is defined
        action_values = []
        
        # Check value of taking every possible action
        for a in range(env.n_actions):
            q_val = 0
            for prob, s_prime, reward, done in env.transitions[s][a]:
                future_val = 0 if done else V[s_prime]
                q_val += prob * (reward + gamma * future_val)
            action_values.append(q_val)
        
        # Choose best action (Greedy)
        best_action = np.argmax(action_values)
        new_policy[s] = best_action
        
        if old_action != best_action:
            policy_stable = False
            
    return new_policy, policy_stable

def run_policy_iteration():
    env = GridWorld()
    
    # Initialization
    policy = np.zeros(env.n_states, dtype=int) 
    V = np.zeros(env.n_states)
    
    iteration = 0
    while True:
        iteration += 1
        print(f"--- Iteration {iteration} ---")
        
        # 1. Evaluate current policy
        V = policy_evaluation(env, policy, V)
        
        # 2. Improve policy
        # <--- We now pass 'policy' to the function
        policy, stable = policy_improvement(env, V, policy) 
        
        # Visualization
        print(f"Policy Stable: {stable}")
        if stable:
            print("Converged! Optimal Policy found.")
            break
            
    return V, policy, env

# Run it
V_opt, p_opt, env = run_policy_iteration()
```

```python
# ==========================================
# 4. PLOTTING RESULTS
# ==========================================
def plot_results(V, policy, size):
    # Map actions to arrows for display
    arrows = {0: '↑', 1: '→', 2: '↓', 3: '←'}
    policy_grid = np.array([arrows[a] for a in policy]).reshape(size, size)
    
    # Heatmap of Values
    plt.figure(figsize=(8, 6))
    sns.heatmap(V.reshape(size, size), annot=True, cmap="YlGnBu", fmt=".3f")
    
    # Overlay arrows
    for i in range(size):
        for j in range(size):
            action_arrow = policy_grid[i, j]
            # Don't plot arrows on Goal or Holes
            state_idx = i * size + j
            if state_idx == env.goal_state: text = "GOAL"
            elif state_idx in env.holes: text = "HOLE"
            else: text = action_arrow
            
            plt.text(j + 0.5, i + 0.5, text, 
                     ha='center', va='center', color='red', fontsize=16, fontweight='bold')
            
    plt.title("Optimal Value Function & Policy Map")
    plt.show()

plot_results(V_opt, p_opt, env.size)
```

```python
import numpy as np
import matplotlib.pyplot as plt

class GridWorld(object):
    def __init__(self, m, n, magicSquares):
        self.grid = np.zeros((m,n))
        self.m = m
        self.n = n
        self.stateSpace = [i for i in range(self.m*self.n)]
        self.stateSpace.remove(80)
        self.stateSpacePlus = [i for i in range(self.m*self.n)]
        self.possibleActions = ['U', 'D', 'L', 'R']
        self.actionSpace = {'U': -self.m, 'D': self.m,
                            'L': -1, 'R': 1}
        self.P = {}
        # dict with magic squares and resulting squares
        self.magicSquares = magicSquares
        self.initP()

    def initP(self):
        for state in self.stateSpace:
            for action in self.possibleActions:
                reward = -1
                state_ = state + self.actionSpace[action]
                if state_ in self.magicSquares.keys():
                    state_ = self.magicSquares[state_]
                if self.offGridMove(state_, state):
                    state_ = state
                if self.isTerminalState(state_):
                    reward = 0
                self.P[(state_, reward, state, action)] = 1

    def isTerminalState(self, state):
        return state in self.stateSpacePlus and state not in self.stateSpace

    def offGridMove(self, newState, oldState):
        # if we move into a row not in the grid
        if newState not in self.stateSpacePlus:
            return True
        # if we're trying to wrap around to next row
        elif oldState % self.m == 0 and newState  % self.m == self.m - 1:
            return True
        elif oldState % self.m == self.m - 1 and newState % self.m == 0:
            return True
        else:
            return False

def printV(V, grid):
    for idx, row in enumerate(grid.grid):
        for idy, _ in enumerate(row):
            state = grid.m * idx + idy
            print('%.2f' % V[state], end='\t')
        print('\n')
    print('--------------------')

def printPolicy(policy, grid):
    for idx, row in enumerate(grid.grid):
        for idy, _ in enumerate(row):
            state = grid.m * idx + idy
            if not grid.isTerminalState(state):
                if state not in grid.magicSquares.keys():
                    print('%s' % policy[state], end='\t')
                else:
                    print('%s' % '--', end='\t')
            else:
                print('%s' % '--', end='\t')
        print('\n')
    print('--------------------')

def evaluatePolicy(grid, V, policy, GAMMA, THETA):
    # policy evaluation for the random choice in gridworld
    converged = False
    i = 0
    while not converged:
        DELTA = 0
        for state in grid.stateSpace:
            i += 1
            oldV = V[state]
            total = 0
            weight = 1 / len(policy[state])
            for action in policy[state]:
                for key in grid.P:
                    (newState, reward, oldState, act) = key
                    # We're given state and action, want new state and reward
                    if oldState == state and act == action:
                        total += weight*grid.P[key]*(reward+GAMMA*V[newState])
            V[state] = total
            DELTA = max(DELTA, np.abs(oldV-V[state]))
            converged = True if DELTA < THETA else False
    print(i, 'sweeps of state space in policy evaluation')
    return V

def improvePolicy(grid, V, policy, GAMMA):
    stable = True   # V is the value functions
    newPolicy = {}
    i = 0
    for state in grid.stateSpace:
        i += 1
        oldActions = policy[state]
        value = []
        newAction = []
        for action in policy[state]:
            weight = 1 / len(policy[state])
            for key in grid.P:
                (newState, reward, oldState, act) = key
                # We're given state and action, want new state and reward
                if oldState == state and act == action:
                    value.append(np.round(weight*grid.P[key]*(reward+GAMMA*V[newState]), 2))
                    newAction.append(action)
        value = np.array(value)
        best = np.where(value == value.max())[0]
        bestActions = [newAction[item] for item in best]
        newPolicy[state] = bestActions

        if oldActions != bestActions:
            stable = False
    print(i, 'sweeps of state space in policy improvement')
    return stable, newPolicy

def iterateValues(grid, V, policy, GAMMA, THETA):
    converged = False
    i = 0
    while not converged:
        DELTA = 0
        for state in grid.stateSpace:
            i += 1
            oldV = V[state]
            newV = []
            for action in grid.actionSpace:
                for key in grid.P:
                    (newState, reward, oldState, act) = key
                    if state == oldState and action == act:
                        newV.append(grid.P[key]*(reward+GAMMA*V[newState]))
            newV = np.array(newV)
            bestV = np.where(newV == newV.max())[0]
            bestState = np.random.choice(bestV)
            V[state] = newV[bestState]
            DELTA = max(DELTA, np.abs(oldV-V[state]))
            converged = True if DELTA < THETA else False

    for state in grid.stateSpace:
        newValues = []
        actions = []
        i += 1
        for action in grid.actionSpace:
            for key in grid.P:
                (newState, reward, oldState, act) = key
                if state == oldState and action == act:
                    newValues.append(grid.P[key]*(reward+GAMMA*V[newState]))
            actions.append(action)
        newValues = np.array(newValues)
        bestActionIDX = np.where(newValues == newValues.max())[0]
        bestActions = actions[bestActionIDX[0]]
        policy[state] = bestActions
    print(i, 'sweeps of state space for value iteration')
    return V, policy

if __name__ == '__main__':
    # map magic squares to their connecting square
    magicSquares = {18: 54, 63: 14}
    env = GridWorld(9, 9, magicSquares)
    # model hyperparameters
    GAMMA = 1.0
    THETA = 1e-6 # convergence criteria

    V = {}
    for state in env.stateSpacePlus:
        V[state] = 0

    policy = {}
    for state in env.stateSpace:
        # equiprobable random strategy
        policy[state] = env.possibleActions

    V = evaluatePolicy(env, V, policy, GAMMA, THETA)
    printV(V, env)

    stable = False
    while not stable:
        V = evaluatePolicy(env, V, policy, GAMMA, THETA)

        stable, policy = improvePolicy(env, V, policy, GAMMA)

    printV(V, env)

    printPolicy(policy, env)

    # initialize V(s)
    V = {}
    for state in env.stateSpacePlus:
        V[state] = 0

    # Reinitialize policy
    policy = {}
    for state in env.stateSpace:
        policy[state] = [key for key in env.possibleActions]

    # 2 round of value iteration ftw
    for i in range(2):
        V, policy = iterateValues(env, V, policy, GAMMA, THETA)

    printV(V, env)
    printPolicy(policy, env)
```
