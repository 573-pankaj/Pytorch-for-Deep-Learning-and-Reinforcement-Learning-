```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
# Since a and b are 3 × 1 and 1 × 2 matrices their shapes do not match up.
```

```python
a.shape, b.shape     
```

```python
a + b
```

```python
a = torch.tensor([8,8,8]).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
print(a)
print(b)
print(a + b)
```

# 2.1.5 Saving Memory

```python
before = id(Y)
print(before)
Y = Y + X
print(id(Y) == before)
print(id(Y))
```

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

```python
before = id(X)
X += Y
id(X) == before
```
